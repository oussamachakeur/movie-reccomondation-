#Getting the usful features to train

```python
x= df2.iloc[: , [2,3,4,5,6,7,8,10]].values
x
```

#Encoding The Categorical Data

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cols_to_encode = [1,4,6,7]

for col in cols_to_encode:
    x[:, col] = le.fit_transform(x[:, col])
x
```

```python
# convert the x into dataframe
column_names = ['age'	,'gender',	'bmi',	'bloodpressure',	'diabetic',	'children',	'smoker',	'claim']

df2 = pd.DataFrame(x, columns=column_names)
```

```python
corr_matrix = df2.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x[:, [0, 2, 3, 5, 6, 7]] = sc.fit_transform(x[:, [0, 2, 3, 5, 6, 7]])
x
```

#Selecting the features to use in clustering based on its importance


```python
y = x[: , [3,6]]
y
```

# K-Mean algorithme

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(1, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(y)
    inertia.append(km.inertia_)

plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method to Determine Optimal K')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
df2['Cluster'] = kmeans.fit_predict(y)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(y, df2['Cluster'])
print(f"Silhouette Score: {score:.2f}")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_components = pca.fit_transform(y)

pca_df = pd.DataFrame(pca_components, columns=['PC1', 'PC2'])
pca_df['Cluster'] = df2['Cluster']
```

```python
import seaborn as sns

sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2')
plt.title('PCA-Reduced Clustering Visualization')
plt.show()
```

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import seaborn as sns

Z = linkage(y, method='ward')  # 'ward' minimizes variance within clusters
```

```python
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=20, leaf_rotation=45, leaf_font_size=12, show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.axhline(y=150, c='k', ls='--', lw=1)  # horizontal cut to suggest cluster level
plt.show()
```

```python
cluster_labels = fcluster(Z, t=2, criterion='maxclust')

x_df = pd.DataFrame(y)
x_df['Cluster'] = cluster_labels
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(x_df.drop('Cluster', axis=1), x_df['Cluster'])
print(f"Silhouette Score (Hierarchical): {score:.3f}")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_components = pca.fit_transform(x_df.drop('Cluster', axis=1))

pca_df = pd.DataFrame(pca_components, columns=['PC1', 'PC2'])
pca_df['Cluster'] = x_df['Cluster']

sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2')
plt.title('PCA Visualization of Hierarchical Clusters')
plt.show()
```
